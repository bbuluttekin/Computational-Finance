### Calculate simple returns

If you denote by $P_t$ the stock price at the end of month $t$, the simple return is given by:

$$R_{t}=\frac{P_{t}−P_{t−1}}{P_{t−1}},$$
the percentage price difference.
Your task in this exercise is to compute the simple returns for every time point n.

**Instructions:**

* Assign to sbux_ret the simple returns for the Starbucks stock.

**Hint:**

It's easier to use pandas .shift() method which shifts elements to number that passed in to it.

```python
sbux_ret = pd.DataFrame((sbux_df['Adj Close']-sbux_df['Adj Close'].shift(1))/sbux_df['Adj Close'].shift(1))
n = len(sbux_ret)-1
sbux_ret = sbux_ret.loc[1:n]
sbux_ret.columns = ['Returns']
sbux_ret
```

### Compute continuously compounded 1-month returns

Continuously compounded returns can be computed easily by realizing that
$$r_t = ln\frac{P_t}{P_{t-1}} = ln(P_t)-ln(P_{t-1})$$

**Instructions:**
* Compute the continuously compounded returns and assign to sbux_ccret.
* Set the names of the sbux_ccret to the correct dates, as you did in the previous exercise.
* Use the head() function to display the first elements of sbux_ccret.

```python
sbux_ccret = pd.DataFrame(np.log(sbux_df['Adj Close'])-
                          np.log(sbux_df['Adj Close'].shift(1))).loc[1:n]
sbux_ccret.index = sbux_df['Date'].loc[1:n]
sbux_ccret.columns = ['Creturns']
sbux_ccret.head()
```

### Compare simple and continuously compounded returns

**Instructions:**
* Display the simple and continuously compounded returns next to each other for the first time periods. What do you notice?

```python
compare = sbux_ccret.join(sbux_ret, how = 'inner')
compare.head()
```

### Calculate growth of $1 invested in SBUX

**Instructions:**
